```python
import pandas as pd
```

```python
def showdf(df, rows):
    """
    函数用于探索给定的DataFrame。
    
    参数:
    - df: 要探索的DataFrame对象。
    - rows: 要打印的头部行数,默认为5。
    """
    if df.isnull().sum().sum() > 0:
        print(f"\nNaN数量: {df.isnull().sum().sum()}\n")
    print(f"\n前{rows}行的信息 (df.shape): {df.shape}")
    # print(f"列名 (df.columns): {df.columns}\n")
    factor_count = df.shape[1] - 2  # 假设第一列为日期，第二列为股票代码
    print(f"因子数量 (计算方法: 总列数 - 2): {factor_count}")
    # print(df.head())
    # print("\ndf.describe():\n", df.describe())
    # print(f"\n列数据类型 (df.dtypes):\n{df.dtypes}\n")
```
